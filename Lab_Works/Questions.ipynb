```python
#Palindrome code
a = input("Enter a number: ")

if a == a[::-1]:
    print("Palindrome")
else:
    print("Not a Palindrome")
```

```python
#Armstrong number code:
num = input("Enter a number: ")
power = len(num)

total = 0
for d in num:
    total += int(d) ** power

if total == int(num):
    print("Armstrong number")
else:
    print("Not an Armstrong number")
```
